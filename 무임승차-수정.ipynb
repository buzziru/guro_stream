```python
import pandas as pd
filepath = 'C:\Class\미드텀\data\지하철_유무임별_이용현황.xls'
df = pd.read_excel(filepath)
df
```

```python
df.head()
```

```python
df.info()
```

```python
df['유임승차'] = df['유임승차'].str.replace(',', '').astype(int)
df['유임하차'] = df['유임하차'].str.replace(',', '').astype(int)
df['무임승차'] = df['무임승차'].str.replace(',', '').astype(int)
df['무임하차'] = df['무임하차'].str.replace(',', '').astype(int)
```

```python
df = df[~df['호선명'].str.contains('신림선|우이신설선|경부선|경인선|경원선|경의선|경강선|경춘선|\
                               |중앙선|수원선|수인선|과천선|장항선|일산선|안산선|9호선|\
                               9호선2~3단계|공항철도 1호선|분당선')]
```

```python
df.info()
```

```python
df['무임승차율']=df['무임승차']/((df['유임승차'])+(df['무임승차']))
```

```python
df['무임하차율']=df['무임하차']/((df['유임하차'])+(df['무임하차']))
```

```python
ride=df.groupby(['호선명','지하철역'])[['무임승차율', '무임하차율']].mean()
ride
```

```python
ride.sort_values(['무임승차율'],ascending=False).head(50)
```

```python
ride.sort_values(['무임하차율'],ascending=False).head(50)
```
