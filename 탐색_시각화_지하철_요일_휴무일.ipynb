```python
import pandas as pd
filepath = '../data/지하철/전처리_지하철/지하철_일별합계_요일_주말.csv'
df = pd.read_csv(filepath, encoding='utf-8', index_col=0)
df.head()
```

```python
ndf = df.copy()
```

```python
data = ndf.loc[:,'승하차구분':'휴무일']
```

```python
df_day = data.groupby('요일').sum()
df_day
```

```python
df_day = df_day.loc[['Mon','Tue','Wed','Thu','Fri','Sat','Sun'],:]
df_day
```

```python
import matplotlib.pyplot as plt
```

## '호선', '역명', '승하차구분'에 따라 그룹화하고 승차, 하차 구분에 따른 이용인원 차이 확인

```python
ride_df = ndf.groupby(['호선', '역명', '승하차구분'], as_index=False).sum()
ride_df
```

```python
mask_up = ride_df[ride_df['승하차구분'] == '승차']
mask_down = ride_df[ride_df['승하차구분'] == '하차']

# ride_df.nlargest('')
```

## 승차인원 상위 10개, 하차인원 상위 10개 필터링

```python
up_10 = mask_up.nlargest(10, '총 이용인원')
up_10.head()
```

```python
down_10 = mask_down.nlargest(10, '총 이용인원')
down_10.head()
```

```python
df_line = ndf.groupby('호선')['총 이용인원'].sum().to_frame()
df_line
```

```python
plt.style.use('ggplot') 
plt.figure(figsize=(14, 7))

line_colors = {
    '1호선': '#0052A4',
    '2호선': '#00A84D',
    '3호선': '#EF7C1C',
    '4호선': '#00A4E3',
    '5호선': '#996CAC',
    '6호선': '#CD7C2F',
    '7호선': '#747F00',
    '8호선': '#E6186C'
}
# RGB 값을 matplotlib에서 사용하는 색상코드로 변환하는 함수
def color_rgb(hex_code):
    hex_code = hex_code.lstrip('#')
    return tuple(int(hex_code[i:i+2], 16)/255 for i in (0, 2, 4))

for i in range(len(df_line)):
    line_name = df_line.index[i]
    plt.bar(df_line.index[i], df_line['총 이용인원'][i], color=color_rgb(line_colors[line_name]))

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('호선별 이용인원(22.05~23.04)', size=35)
plt.xlabel('호    선    명', size=25)
plt.ylabel('이용인원(단위:억명)', size=25)

plt.tight_layout()
```
