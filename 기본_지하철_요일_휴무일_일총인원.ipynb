### 결측치 처리 - 24이후 이용자 없음으로 추정되어 0으로 결측치 처리

```python
ndf = ndf.fillna(0)
ndf.isnull().sum()
```

```python
ndf['수송일자'] = pd.to_datetime(ndf['수송일자'])
ndf['수송일자']
```

```python
ndf['요일'] = ndf['수송일자'].dt.weekday
ndf
```

```python
ndf.loc[ndf['요일'].isin([5, 6]), '휴무일'] = '주말'
ndf.loc[ndf['요일'].isin([0,1,2,3,4]), '휴무일'] = '주중'
ndf
```

```python
ndf['휴무일'].value_counts() # 공휴일도 확인해야 함
```

```python
ndf['24시이후'].astype('int')
```

```python
ndf['총 이용인원'] = ndf.loc[:,'06시이전':'24시이후'].sum(axis=1).astype('int')
ndf
```

```python
ndf.columns
```

```python
col = ['수송일자', '호선', '고유역번호(외부역코드)', '역명', '승하차구분', '06시이전', '06-07시간대',
       '07-08시간대', '08-09시간대', '09-10시간대', '10-11시간대', '11-12시간대', '12-13시간대',
       '13-14시간대', '14-15시간대', '15-16시간대', '16-17시간대', '17-18시간대', '18-19시간대',
       '19-20시간대', '20-21시간대', '21-22시간대', '22-23시간대', '23-24시간대', '24시이후','총 이용인원',
       '요일', '휴무일']
ndf = ndf.loc[:,col]
ndf
```

```python
ndf['요일'] = ndf['수송일자'].dt.strftime('%a')
ndf
```
