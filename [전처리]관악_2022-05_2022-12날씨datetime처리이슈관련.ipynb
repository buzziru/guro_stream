방법을 하기와 같이 바꿔 사용

```python
# 실제 Null값에 대한 시간별 온도를 확인하여 하기와 같이 반영 (hourly values)
real_data = [23.9, 23.8, 23.6, 23.5, 23.3, 23.3, 22.9, 22.9, 22.4, 22.4, 22.6, 22.6, 23.0, 22.6]

# null값 시작과 끝 시간 반영
start_time = '2022-08-02 02:00:00'
end_time = '2022-08-02 13:00:00'

# Create a list of hourly timestamps for the specified datetime range
hourly_timestamps = pd.date_range(start=start_time, end=end_time, freq='H')

# Create a list of minute timestamps within the specified hourly range
minute_timestamps = pd.date_range(start=start_time, end=end_time, freq='T')

# Filter the DataFrame 'ndf' to include only the rows within the specified datetime range
filtered_ndf = ndf[(ndf['일시'] >= start_time) & (ndf['일시'] <= end_time)].copy()

# Repeat the 'real_data' values for each minute within the specified hourly range
repeated_real_data = [real_data[i // 60] for i in range(len(minute_timestamps))]

# Update the '기온(°C)' column in the filtered DataFrame with the repeated real data values
filtered_ndf['기온(°C)'] = repeated_real_data[:len(filtered_ndf)]

# Update the corresponding rows in the original DataFrame 'ndf'
ndf.update(filtered_ndf)


ndf
```

누적강수량의 null값은 0으로 처리
기온의 나머지 미미한 null값은 fillna 함수의 ffill옵션으로 처리

```python
ndf['누적강수량(mm)'].fillna(0, inplace=True)
ndf['기온(°C)'].fillna(method = 'ffill', inplace=True)
ndf
```

```python
ndf['일시'] = pd.to_datetime(ndf['일시'])
ndf['time_str'] = ndf['일시'].dt.strftime('%Y-%m-%d %H:%M:%S')
ndf = ndf.set_index('time_str')
ndf
```

```python
ndf.loc['2022-01-01 00:01:00':'2022-04-30 23:59:00']
```

```python
ndf = ndf.drop(ndf.loc['2022-01-01 00:01:00':'2022-04-30 23:59:00'].index)
```

```python
ndf_group = ndf.groupby(ndf['일시'].dt.floor('H')).agg({'기온(°C)':'mean', '누적강수량(mm)':'max', '습도(%)':'mean'})

ndf_group['시간당 강수량(mm)'] = ndf_group['누적강수량(mm)'].diff()
ndf_group['시간당 강수량(mm)'].fillna(0, inplace=True)
ndf_group['시간당 강수량(mm)'][ndf_group['시간당 강수량(mm)'] < 0] = 0
ndf_group
```
