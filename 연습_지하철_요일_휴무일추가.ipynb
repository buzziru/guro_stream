```python
import pandas as pd
filepath = r'C:\CIG\data\전처리_지하철 시간대 승하차(일별)\서울교통공사_역별 일별 시간대별 승하차인원 정보 - 출퇴근시간.csv'
df = pd.read_csv(filepath, encoding='ANSI')
```

```python
ndf = df.copy()
```

```python
ndf = ndf.set_index('연번')
ndf
```

```python
ndf.info()
```

```python
ndf['수송일자'] = pd.to_datetime(ndf['수송일자'])
ndf['수송일자']
```

```python
ndf['요일'] = ndf['수송일자'].dt.weekday
ndf
```

```python
ndf['요일'].dtype
```

```python
ndf.loc[ndf['요일'].isin([5, 6]), '주말'] = 1
ndf.loc[ndf['요일'].isin([0,1,2,3,4]), '주말'] = 0
ndf
```
