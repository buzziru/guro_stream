```python
import pandas as pd
```

```python
file_path = r'C:\CIG\data\지하철 시간대별 승하차(일별)\서울교통공사_역별 일별 시간대별 승하차인원 정보_20230430.csv'
df = pd.read_csv(file_path, encoding='ANSI')
df.head()
```

#### 모든 열에 대해 결측치 없음 확인. '수송일자' dtype 변환

```python
df.info()
```

```python
ndf = df.copy()
```

```python
ndf['수송일자'] = pd.to_datetime(ndf['수송일자'])
```

```python
ndf['수송일자'].dtype
```

# '수송일자' 열을 이용하여 요일 구하기

```python
ndf['요일'] = ndf['수송일자'].dt.weekday
ndf.head()
```

```python
ndf.columns
```

# '요일'을 보기 좋게 '수송일자' 다음으로 정렬

```python
col = ['연번', '수송일자','요일', '호선', '역번호', '역명', '승하차구분', '06시이전', '06-07시간대',
       '07-08시간대', '08-09시간대', '09-10시간대', '10-11시간대', '11-12시간대', '12-13시간대',
       '13-14시간대', '14-15시간대', '15-16시간대', '16-17시간대', '17-18시간대', '18-19시간대',
       '19-20시간대', '20-21시간대', '21-22시간대', '22-23시간대', '23-24시간대', '24시이후'
       ]
ndf_renamed = ndf.loc[:, col]
ndf_renamed
# 0은 월요일, 6은 일요일
```

### 숫자로 된 요일을 문자열로 변환

```python
ndf_renamed['요일'] = ndf_renamed['수송일자'].dt.strftime('%a')
ndf_renamed
```

```python
ndf_renamed.to_csv(r'C:\CIG\data\전처리_지하철 시간대 승하차(일별)\지하철_요일추가_2023.csv')
```
