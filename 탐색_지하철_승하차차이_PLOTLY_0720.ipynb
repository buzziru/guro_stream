```python
import plotly.express as px
```

```python
import pandas as pd

df = pd.read_csv('../data/지하철/전처리_지하철/전처리_지하철_요일_고유역번호_승하차차이_위도경도0717.csv', index_col=0)
```

```python
df.head()
```

```python
diff = df[df['승하차구분']=='승하차 차이']
diff
```

```python
data = diff.groupby(['호선','역명','요일']).sum().loc[:,'06시이전':'총 이용인원']
data
```

```python
data = data.reset_index()
data
```

```python
guro = data[data['역명'] == '구로디지털단지']
guro
```

```python
guro_rdy = guro.drop(['호선','역명'], axis=1)
guro_rdy
```

```python
guro_time = guro_rdy['06시이전':'24시이후']
```

```python
guro_rdy = guro_rdy.set_index('요일')
guro_rdy
```

```python
guro_t = guro_rdy.T
guro_t
```

```python
guro.columns[3:]
```

```python
fig = px.line(guro_t, x=guro_t.index, y=0.0, labels={"index":'시    간', "0.0":'승하차 차이'},
              title='구디역 시간별 승하차 차이', markers=True)   

fig.update_layout(title_x=0.5)
fig
```

```python
guro_time.head()
```

```python
fig = px.line(guro_time, x=guro_t.columns, y=0.0, labels={"index":'시    간', "0.0":'승하차 차이'},
              title='구디역 시간별 승하차 차이', markers=True, colors=guro_time.index)   

fig.update_layout(title_x=0.5)
fig
```
