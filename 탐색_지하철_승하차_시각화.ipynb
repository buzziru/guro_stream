```python
import folium
import numpy as np
from ast import literal_eval
```

```python
# 호선별 월요일 출근시간 승하차 차이
max_pop = max(result_diff['08-09시간대'][result_diff['요일']==0])

for i in range(1, 9):
    seoul_map = folium.Map(location=[37.55,126.98], zoom_start=13) 
    line_name = f'{i}호선'    
    for index, row in line[i][line[i]['요일']==0].iterrows():
        location = literal_eval(row['location'])        
        color = 'blue' if row['08-09시간대'] >=0 else 'red'        
        radius = abs(row['08-09시간대']/max_pop) * 40
        popup = folium.Popup(row['역명'], max_width=200)        
        folium.CircleMarker(location=location, popup=popup,\
                            color=color, fill=True,fill_opacity=0.6,\
                            radius=radius).add_to(seoul_map)
    seoul_map.save(f'./map_visual/seoul_diff_{i}_mon_08.html')
```
