# 호선 별로 지도에 마커생성

### 1) 기본 아이콘

```python
from folium.features import CustomIcon

for j in range(1, 9):
    seoul_map = folium.Map(location=[37.55,126.98], zoom_start=13)
    for index, row in subways[j].iterrows():
        popup = folium.Popup(row['역명'], max_width=200)
        folium.Marker(location=row['location'], popup=popup).add_to(seoul_map)
    seoul_map.save(f'./map_visual/seoul_{j}.html')
```

### 2) 호선 별로 아이콘 지정

```python
from folium.features import CustomIcon

for j in range(1, 9):
    seoul_map = folium.Map(location=[37.55,126.98], zoom_start=13)
    for index, row in subways[j].iterrows():

        # 호선 별로 다른 아이콘 지정
        icon_image = f'./icons/icon_{j}.svg'
        icon = CustomIcon(icon_image=icon_image, icon_size=(30, 30))
        
        popup = folium.Popup(row['역명'], max_width=200)
        folium.Marker(location=row['location'], popup=popup, icon=icon).add_to(seoul_map)
    seoul_map.save(f'./map_visual/seoul_{j}.html')
```

### 3) 호선 별로 아이콘 지정 - 로컬 서버 문제 해결

```python
from folium.features import CustomIcon
import base64

for j in range(1, 9):
    seoul_map = folium.Map(location=[37.55,126.98], zoom_start=13)

    for index, row in subways[j].iterrows():

        with open(f'./icons/icon_{j}.svg', 'r') as f:
            svg = f.read()
        encoded = base64.b64encode(svg.encode('utf-8')).decode('utf-8')
        icon_image = f'data:image/svg+xml;base64,{encoded}' 
        icon = folium.CustomIcon(icon_image=icon_image, icon_size=(30, 30))

        popup = folium.Popup(row['역명'], max_width=200)
        folium.Marker(location=row['location'], popup=popup, icon=icon).add_to(seoul_map)
    seoul_map.save(f'./map_visual/seoul_{j}.html')
```

## 병합된 데이터 프레임을 호선별로 분리

```python
result1 = total.pivot_table(index=['호선', '역명', '승하차구분', 'location'],\
                  values=['08-09시간대', '09-10시간대', '18-19시간대', '19-20시간대','총 이용인원'], aggfunc='sum')
result1
```

```python
result2 = total.pivot_table(index=['호선', '역명', 'location'],\
                  values=['08-09시간대', '09-10시간대', '18-19시간대', '19-20시간대','총 이용인원'], aggfunc='sum')
result2
```

```python
result2 = result2.reset_index()
```

```python
result2
```

```python
line = {}
for i in range(1, 9):
    line[i] = result2[result2['호선'] == f'{i}호선']
line[2].head()
```

```python
line[2].info()
```

```python
line_colors = {
    '1호선': '#0052A4',
    '2호선': '#00A84D',
    '3호선': '#EF7C1C',
    '4호선': '#00A4E3',
    '5호선': '#996CAC',
    '6호선': '#CD7C2F',
    '7호선': '#747F00',
    '8호선': '#E6186C'
}
max_pop = max(max(line[i]['총 이용인원']) for i in range(1, 9))

for i in range(1, 9):
    seoul_map = folium.Map(location=[37.55,126.98], zoom_start=13) 
    line_name = f'{i}호선'
    for index, row in line[i].iterrows():        
        popup = folium.Popup(row['역명'], max_width=200)
        radius = 70 * (row['총 이용인원']/max_pop)
        folium.CircleMarker(location=row['location'], popup=popup,\
                            color=line_colors[line_name], fill_color=line_colors[line_name],\
                            radius=radius).add_to(seoul_map)
    seoul_map.save(f'./map_visual/seoul_result{i}.html')
```

```python
line_colors = {
    '1호선': '#0052A4',
    '2호선': '#00A84D',
    '3호선': '#EF7C1C',
    '4호선': '#00A4E3',
    '5호선': '#996CAC',
    '6호선': '#CD7C2F',
    '7호선': '#747F00',
    '8호선': '#E6186C'
}
max_pop_tot = max(max(result2['총 이용인원']) for i in result2.iterrows())

seoul_map = folium.Map(location=[37.55,126.98], zoom_start=13)
for index, row in result2.iterrows():
    line_name = row['호선']
    popup = folium.Popup(row['역명'], max_width=200)
    radius = 50 * (row['총 이용인원']/max_pop_tot)
    folium.CircleMarker(location=row['location'], popup=popup,\
                        color=line_colors[line_name], fill_color=line_colors[line_name],\
                        radius=radius).add_to(seoul_map)
    seoul_map.save(f'./map_visual/seoul_result_tot.html')
```
