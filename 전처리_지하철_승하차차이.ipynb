```python
import pandas as pd
filepath = '../data/지하철/전처리_지하철/전처리_지하철_요일_고유역번호0717.csv'
df = pd.read_csv(filepath, encoding='utf-8', index_col=0)
df.head()
```

```python
ndf = df.copy()
```

```python
df_up = ndf[ndf['승하차구분'] == '승차']
df_down = ndf[ndf['승하차구분'] == '하차']

df_diff = df_up.set_index(['수송일자', '호선', '고유역번호(외부역코드)', '역명']).loc[:,'06시이전':'총 이용인원']\
- df_down.set_index(['수송일자', '호선', '고유역번호(외부역코드)', '역명']).loc[:,'06시이전':'총 이용인원']

df_diff.head()
```

```python
df_diff['승하차구분'] = '승하차 차이'
df_diff.head()
```

```python
df_diff.reset_index(inplace=True)
```

```python
df_diff.isnull().sum()
```

```python
result = pd.concat([ndf, df_diff])
result.tail()
```

```python
result = result.sort_values(by=['수송일자', '호선', '고유역번호(외부역코드)', '역명'])
result
```

```python
result.reset_index(drop=True, inplace=True)
result
```

```python
result.info()
```

```python
result['수송일자'] = pd.to_datetime(result['수송일자'])
result['수송일자']
```

```python
result['요일'] = result['수송일자'].dt.weekday
result.loc[result['요일'].isin([5, 6]), '휴무일'] = 1
result.loc[result['요일'].isin([0,1,2,3,4]), '휴무일'] = 0
result.info()
```

```python
result.to_csv('../data/지하철/전처리_지하철/전처리_지하철_요일_고유역번호_승하차차이0717.csv')
```
