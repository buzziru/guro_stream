```python
ndf['시간당 강수량(mm)'] = ndf['시간당 강수량(mm)'].fillna(0)
```

```python
ndf[ndf['시간당 강수량(mm)'] > 0]
```

```python
ndf.drop(columns=['누적강수량(mm)', '강수량차이'], inplace=True)
ndf
```

```python
def rain(x):
    if x == 0:
        return '해당 없음'
    elif x < 3:
        return '약한 비'
    elif x < 15:
        return '보통 비'
    elif x < 30:
        return '강한 비'
    elif x >=30:
        return '매우 강한 비'    
```

```python
final_data['강수량 분류'] = final_data['시간당 강수량(mm)'].apply(rain)
final_data
```

```python
final_data['강수량 분류'].value_counts()
```

```python
# 폭염 분류 함수 정의
def heat_wave(x):
    heat_dates = [
        '2022-07-02',
        '2022-07-03',
        '2022-07-05',
        '2022-07-06',
        '2022-07-10',
        '2022-07-26',
        '2022-07-27',
        '2022-07-28',
        '2022-07-29',
        '2022-07-30',
    ]
    if x.strftime('%Y-%m-%d') in heat_dates:
        return 'True'
    else:
        return 'False'
```

```python
# '폭염 분류' 열에 함수 적용하여 값 할당
final_data['폭염 분류'] = final_data.index.map(heat_wave)
final_data
```

```python
final_data[final_data['폭염 분류']=='True']
```

```python
# 한파 분류 함수 정의
def cold_wave(x):
    cold_dates = [
        '2022-12-18',
        '2022-12-19',
        '2022-12-23',
        '2022-12-24',
        '2023-01-24',
        '2023-01-25',
    ]
    if x.strftime('%Y-%m-%d') in cold_dates:
        return 'True'
    else:
        return 'False'
    
#참조: 한파일수: 아침 최저기온(03:01~09:00)이 영하 12도 이하인 날의 수 의미_기상청 기상자료개방포털
```

```python
# '폭염 분류' 열에 함수 적용하여 값 할당
final_data['한파 분류'] = final_data.index.map(cold_wave)
final_data
```
